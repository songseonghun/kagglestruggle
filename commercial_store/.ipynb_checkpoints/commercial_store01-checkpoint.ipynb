```python
shop_201809 = pd.read_csv('../data/commercial_store/소상공인시장진흥공단_상가업소정보_201809_1.csv', engine='python', encoding='cp949')
shop_201809.shape
```

```python
view_columns = ['상호명', '지점명', '상권업종대분류명', '상권업종중분류명',
              '상권업종소분류명', '시도명', '시군구명', 
              '행정동명', '법정동명', '지번주소', '경도', '위도']
shop_201809[view_columns].head()
```

```python
#결측치 보기
shop_201809.isnull().sum()
```

```python
import missingno as msno

msno.matrix(shop_201809)
```

```python
(ggplot(shop_201809[:1000])
    + aes(x='경도',y="위도")
    + geom_point()
    + theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
#서울과 부산만 데이터가 있는 것으로 추정되어, 서울만 데이터를 나눔

shop_seoul = shop_201809.loc[shop_201809['도로명주소'].str.startswith('서울')]
shop_except_seoul = shop_201809.loc[~shop_201809['도로명주소'].str.startswith('서울')]
print(shop_seoul.shape)
print(shop_except_seoul.shape)
```

```python
(ggplot(shop_seoul)
    +aes(x="경도", y="위도")
    +geom_point(color='black', alpha=0.2, size=0.2)
    +theme(text=element_text(family='NanumBarunGothic'))
)
```

```python
(ggplot(shop_except_seoul)
    +aes(x="경도", y="위도")
    +geom_point(color='black', alpha=0.2, size=0.2)
    +theme(text=element_text(family='NanumBarunGothic'))
)
```
